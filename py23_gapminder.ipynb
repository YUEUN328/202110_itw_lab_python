* GitHub에 저장된 [gapminder.tsv](https://github.com/JakeOh/202110_itw_lab_python/raw/main/gapminder.tsv) 파일을 읽고 DataFrame을 생성
* 데이터 프레임 탐색
* (중복되지 않는) 국가 이름의 개수?
* (중복되지 않는) 대륙 이름?
* (중복되지 않는) 연도의 개수?
* 국가이름, 연도, 1인당 GDP 컬럼을 선택
* 1인당 GDP의 최댓값, 최솟값을 찾고, 그 row를 출력
* 기대수명(lifeExp)의 최솟값과 최댓값을 찾고, 그 row를 출력

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# GitHub에 저장된 파일 경로
file_path = 'https://github.com/JakeOh/202110_itw_lab_python/raw/main/gapminder.tsv'
```

```python
# CSV(Comma-Separated Values), TSV(Tab-Separated Values)
gapminder = pd.read_csv(file_path, sep = '\t')
```

```python
gapminder.info()
```

```python
gapminder.head()
```

```python
gapminder.tail()
```

```python
# 숫자 타입 변수들의 기술 통계량
gapminder.describe()
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(10, 4))

sns.boxplot(y='lifeExp', data=gapminder, ax=ax[0])
sns.boxplot(y='pop', data=gapminder, ax=ax[1])
sns.boxplot(y='gdpPercap', data=gapminder, ax=ax[2])

plt.show()
```

```python
# 나라 이름 개수
len(gapminder['country'].unique())
```

```python
# 대륙 이름 개수
len(gapminder['continent'].unique())
```

```python
# 연도 개수
len(gapminder['year'].unique())
```

```python
gapminder.columns
```

```python
# fancy indexing - 여러 개의 컬럼을 함께 선택하는 방법
gapminder[['country', 'year', 'gdpPercap']]
```

```python
# 1인당 GDP 최솟값
gapminder['gdpPercap'].min()
```

```python
# 1인당 GDP 최댓값
gapminder['gdpPercap'].max()
```

```python
# 1인당 GDP 최솟값인 행과 최댓값인 행 - boolean indexing(&, |, ~)
gapminder[(gapminder['gdpPercap'] == gapminder['gdpPercap'].min()) | 
          (gapminder['gdpPercap'] == gapminder['gdpPercap'].max())]
```

```python
# 기대 수명 최솟값, 최댓값인 행
gapminder[(gapminder['lifeExp'] == gapminder['lifeExp'].min()) | 
          (gapminder['lifeExp'] == gapminder['lifeExp'].max())]
```

```python
# Japan의 기대 수명 연도별 그래프
japan = gapminder[gapminder['country'] == 'Japan']
japan
```

```python
sns.lineplot(x='year', y='lifeExp', data=japan, marker='o')
plt.grid()
plt.show()
```

```python
rwanda = gapminder[gapminder['country'] == 'Rwanda']
rwanda
```

```python
sns.lineplot(x='year', y='lifeExp', data=rwanda, marker='o')
plt.grid()
plt.show()
```

```python
# 연도별 기대 수명의 평균
gapminder.groupby(by='year')['lifeExp'].mean()
```

```python
sns.lineplot(x='year', y='lifeExp', data=gapminder, 
             marker='o', err_style=None)
plt.grid()
plt.show()
```

```python
sns.barplot(x='year', y='lifeExp', data=gapminder)
plt.show()
```

```python
# 연도별 1인당 GDP 평균
gapminder.groupby(by='year')['gdpPercap'].mean()
```

```python
sns.lineplot(x='year', y='gdpPercap', data=gapminder, marker='o')
plt.grid()
plt.show()
```

```python
sns.barplot(x='year', y='gdpPercap', data=gapminder)
plt.show()
```

```python
# 연도별 대륙별 전세계 인구
gapminder.groupby(by=['year', 'continent'])['pop'].sum()
```

```python
sns.lineplot(x='year', y='pop', hue='continent', data=gapminder, 
             estimator=np.sum, marker='o', err_style=None)
plt.grid()
plt.show()
```

* 나라 이름에 'Korea'가 포함된 부분집합

```python
gapminder[gapminder['country'] == 'Korea']
```

```python
# 나라 이름에 'Korea'가 포함된 부분집합
gapminder[gapminder['country'].str.contains('Korea')]
```

```python
gapminder[gapminder['country'].str.contains('korea')]
#> 영문 대/소문자를 구분하기 때문에 조건을 만족하는 행이 없음.
```

```python
# 나라 이름을 모두 소문자로 변환한 후 'korea'를 포함하고 있는지를 검사
gapminder[gapminder['country'].str.lower().str.contains('korea')]
```

```python
korea = gapminder[gapminder['country'].str.lower().str.contains('korea, rep')]
korea
```

```python
# 우리나라 기대 수명 시계열 그래프
sns.lineplot(x='year', y='lifeExp', data=korea, marker='o')
plt.grid()
plt.show()
```
